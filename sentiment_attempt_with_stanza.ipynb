```python
import pandas as pd
import stanza
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
nlp = stanza.Pipeline(lang='en', processors='tokenize,sentiment', tokenize_no_ssplit=True)
```

```python
df = pd.read_csv("/content/drive/MyDrive/Royal mail Data ratings.csv")
```

```python
df['reviewBody'] = df['reviewBody'].str.lower()
```

```python
def get_sentiment(review):
    doc = nlp(review)
    sentiment_label = doc.sentences[0].sentiment
    return 0 if sentiment_label == 'positive' else 1
```

```python
df['sentiment'] = df['reviewBody'].apply(lambda x: get_sentiment(x))
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['reviewBody'], df['sentiment'], test_size=0.2, random_state=42)
```
